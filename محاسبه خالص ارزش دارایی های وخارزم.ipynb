```python
price = [1438,15810,2913,6250,4147,4792,2154,1427,16250,6830,35760,4476,8640,1495,2894,4039,1737,371,16840,1012]
STOCK_MARKET['price']= price
```

```python
Total_added_value = ((STOCK_MARKET['price'] - STOCK_MARKET['Cost per share'])* STOCK_MARKET['Number of shares']) / 1000000
STOCK_MARKET['Total added value']= Total_added_value
```

جمع ارزش افزوده پرتفوی بورسی  

```python
Sum__of_stock_market_added_value = STOCK_MARKET['Total added value'].sum()
Sum__of_stock_market_added_value
```

 محاسبه ستون درصد سهام نسب به کل پرتفوی

```python
sum_cost = STOCK_MARKET['Total cost'].sum()
Percentage_to_the_total_stock_market_portfolio = (STOCK_MARKET['Total cost']/sum_cost)*100
STOCK_MARKET['Percentage to the total stock market portfolio']= Percentage_to_the_total_stock_market_portfolio
STOCK_MARKET
```

محاسبه ارزش روز هر سهم غیر بورسی 

```python
P = 6
Non_stock_portfolio['P/E'] = P
```

```python
Non_stock_portfolio['Daily value of each share'] = Non_stock_portfolio.apply(lambda row: row['Total cost'] if row['EPS'] == 0 else row['P/E'] * row['EPS'], axis=1)
```

محاسبه ارزش افزوده پرتفوی غیر بورسی 

```python
Surplus_value_of_non_stock_portfolio = ((Non_stock_portfolio['Daily value of each share'] - Non_stock_portfolio['Cost per share'])* Non_stock_portfolio['Number of shares']) / 1000000
Non_stock_portfolio['Surplus value of non-stock portfolio']= Surplus_value_of_non_stock_portfolio
Non_stock_portfolio
```

```python
stock_market = pd.DataFrame({'NAV': ['Total added value of stock market'], 'Amount': [STOCK_MARKET['Total added value'].sum()]})
non_stock_market = pd.DataFrame({'NAV': ['Surplus value of non stock portfolio'], 'Amount': [Non_stock_portfolio['Surplus value of non-stock portfolio'].sum()]})
Stored = pd.DataFrame({'NAV': ['Stored depreciation'], 'Amount': [Stored_depreciation]})
dividends = pd.DataFrame({'NAV': ['Total dividends'], 'Amount': [dividends]})
Profit =  pd.DataFrame({'NAV': ['Profit from the sale of shares'], 'Amount': [Profit_from_the_sale_of_shares]})
NAV = pd.concat([NAV, stock_market, non_stock_market], axis=0)
NAV = pd.concat([NAV, Stored], axis=0)
NAV = pd.concat([NAV, dividends], axis=0)
NAV = pd.concat([NAV,Profit], axis=0)
NAV.index = range(len(NAV))


```
